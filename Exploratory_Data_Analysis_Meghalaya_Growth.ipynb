```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load Data (Replace with your actual dataset)
data = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'GSDP': [0.25, 0.27, 0.3, 0.33, 0.37, 0.33, 0.37, 0.41, 0.56, 0.52],
    'Revenue Expenditure': [6252, 6348, 8337, 8423, 10256, 9565, 11499, 13620, 14864, 16570],
    'Capital Expenditure': [1377, 1607, 1835, 1435, 1903, 1402, 2163, 3526, 3362, 5234],
    'State GST Revenue': [340, 350, 360, 376, 806, 910, 823, 1118, 1477, 1723],
    'State Excise Duties Revenue': [151, 170, 169, 199, 226, 276, 375, 308, 365, 458]
})

# Define Features (X) and Target (Y)
X = data[['Revenue Expenditure',
          'Capital Expenditure', 'State GST Revenue', 'State Excise Duties Revenue']]
y = data['GSDP']

# Normalize Features (Z-score standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert to NumPy arrays for plotting
X_array = X.values
y_array = y.values

# Scatter Plots
for col in X.columns:
    plt.figure(figsize=(6, 4))
    plt.scatter(X[col], y, alpha=0.7, edgecolors='k')
    plt.xlabel(col)
    plt.ylabel('GSDP')
    plt.title(f'Scatter Plot: GSDP vs {col}')
    plt.grid(True)

    # Show plot in notebook
    plt.show()

    # Save figure
    plt.savefig(f'{col}_vs_GSDP.png', dpi=300, bbox_inches='tight')

# Heatmap
plt.figure(figsize=(10, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Show in notebook
plt.show()

# Save heatmap
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
```
